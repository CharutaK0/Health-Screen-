```python
df= pd.read_csv(r'C:\Users\Charuta\Codes\Python\Health Screen\Data\unclean enriched_deficiency_dataset.csv')
df.drop(columns= ['Iron_Intake_mg', 'VitaminD_Intake_IU','B12_Intake_mcg', 'Calcium_Intake_mg', 'Magnesium_Intake_mg', 'Zinc_Intake_mg', 'Folate_Intake_mcg', 'Serum_Iron_ugdL', 'Serum_Ferritin_ngmL', 'Hemoglobin_gdL', 'MCV_fL', 'Transferrin_Saturation_pct', 'Serum_VitD_ngmL', 'Serum_B12_pgmL', 'Serum_Folate_ngmL', 'Serum_Calcium_mgdL', 'Serum_Magnesium_mgdL', 'Serum_Zinc_ugdL'], inplace = True)
df.dropna(axis= 'index',how='all') #this is used when you want to drop indexes with all values missing
df.dropna(axis= 'columns', how= 'all')
# Step 1 — Fill known nullable columns with 'None'
cat_null_cols = [
    'Pregnancy_Status', 'Alcohol_Consumption',
    'Chronic_Condition', 'Medication_Use', 'Digestive_Issues'
]
df[cat_null_cols] = df[cat_null_cols].fillna('None')

# Step 2 — Fill any remaining Yes/No columns with 'No'
# Empty symptom = symptom not present = No
binary_null_cols = [
    'Fatigue', 'Pallor', 'Shortness_of_Breath', 'Cold_Intolerance',
    'Bone_Pain', 'Muscle_Weakness', 'Frequent_Illness', 'Hair_Loss',
    'Tingling_Numbness', 'Memory_Cognitive_Issues', 'Balance_Problems',
    'Mood_Swings', 'Muscle_Cramps', 'Dental_Issues', 'Brittle_Nails',
    'Anxiety_and_Poor_Sleep', 'Frequent_Headaches', 'Loss_of_Appetite',
    'Slow_Wound_Healing', 'Loss_of_Taste_Smell', 'Mouth_Sores',
    'Tongue_Swelling', 'Restless_Legs', 'Night_Sweats',
    'Recent_Blood_Loss', 'Family_History_of_Deficiency'
]
df[binary_null_cols] = df[binary_null_cols].fillna('No')

# Step 3 — Drop rows where target is still null
df.dropna(subset=['Deficiency'], inplace=True)

# Step 4 — Final check
print("Remaining nulls:")
print(df.isnull().sum()[df.isnull().sum() > 0])
binary_cols = [
    'Fatigue', 'Pallor', 'Shortness_of_Breath', 'Cold_Intolerance',
    'Bone_Pain', 'Muscle_Weakness', 'Frequent_Illness', 'Hair_Loss',
    'Tingling_Numbness', 'Memory_Cognitive_Issues', 'Balance_Problems',
    'Mood_Swings', 'Muscle_Cramps', 'Dental_Issues', 'Brittle_Nails',
    'Anxiety_and_Poor_Sleep', 'Frequent_Headaches', 'Loss_of_Appetite',
    'Slow_Wound_Healing', 'Loss_of_Taste_Smell', 'Mouth_Sores',
    'Tongue_Swelling', 'Restless_Legs', 'Night_Sweats',
    'Recent_Blood_Loss', 'Family_History_of_Deficiency'
]
for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})
print("Binary Columns done!")
print(df[binary_cols].head)

df['Energy_Level'] = df['Energy_Level'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['Sleep_Quality'] = df['Sleep_Quality'].map({'Poor': 0, 'Average': 1, 'Good': 2})
df['Sun_Exposure'] = df['Sun_Exposure'].map({'Low': 0, 'Moderate': 1, 'High': 2})
df['Stress_Level'] = df['Stress_Level'].map({'Low': 0, 'Moderate': 1, 'High': 2})
df['Physical_Activity_Level'] = df['Physical_Activity_Level'].map({'Sedentary': 0, 'Light': 1, 'Moderate': 2, 'Active': 3})
df['Digestive_Issues']    =df['Digestive_Issues'].map({'None':0, 'Occasional':1, 'Frequent':2})
print("Ordered columns encoded!")

one_hot_cols = [
    'Gender', 'Geographic_Region', 'Diet_Type',
    'Smoking_Status', 'Alcohol_Consumption',
    'Chronic_Condition', 'Medication_Use', 'Pregnancy_Status'
]
df = pd.get_dummies(df, columns=one_hot_cols, drop_first=False)

print("One hot encoding done!")
print("New shape:", df.shape)
```
